```python
import jieba
import jieba.analyse
import json
```

```python
my_str = "如果你是一名资讯工作者，或者是在某个领域有所专长，那么你一定会经常浏览网上的问题和回答。"
```

```python
seg_list = jieba.cut(my_str, cut_all=True)
for word in seg_list:
    print(word,end=' ')
```

```python
seg_list = jieba.cut(my_str, cut_all=False)
for word in seg_list:
    print(word,end=' ')
```

```python
seg_list = jieba.cut_for_search(my_str)
for word in seg_list:
    print(word,end=' ')
```

```python
seg_list = jieba.cut(my_str, cut_all=False) + jieba.lcut(my_str)
for word in seg_list:
    print(word)
```

```python
segs1 = jieba.cut(my_str, cut_all=False)
segs2 = jieba.cut(" ".join(segs1), cut_all=False)
print(" ".join(segs2))
segs = segs1 + segs2
for word in segs:
    print(word)
```

```python
with open('../18697.xml', 'r', encoding='utf-8') as f:
    my_str = f.read()
```

```python
import bs4
soup = bs4.BeautifulSoup(my_str, 'xml')
my_str = soup.find('QW').get('value')
my_str
```

```python
words = jieba.cut_for_search(my_str)
words = [word for word in words if word not in stop_words]
print(words)
json.dump(words, open('18697.json', 'w', encoding='utf-8'), ensure_ascii=False)
```

```python
# 可以看到对人名的分词效果很差，这是因为jieba默认使用的是HMM模型，而HMM模型对于未登录词的识别效果很差。
# 可以结合数据库里存的当事人信息等，对人名进行特殊处理。
jieba.load_userdict('user_dict.txt')
words = jieba.cut_for_search(my_str)
words = [word for word in words if word not in stop_words]
print(words)
```
